```python
sequences = {}
current_name = ''

with open('rosalind_GC.txt', 'r') as f:
    for line in f:
        line = line.strip()
        if line.startswith('>'):
            current_name = line[1:]
            sequences[current_name] = ''
        else:
            sequences[current_name] = sequences[current_name] + line

for name in sequences:
    sequence = sequences[name]
    count_G = sequence.count('G')
    count_C = sequence.count('C')
    GC_content = (count_G + count_C)/len(sequence) * 100
    best_name = ''
    best_GC = 0
    if GC_content > best_GC:
        best_GC = GC_content
        best_name = name

print(best_name)
print(GC_content)
```
