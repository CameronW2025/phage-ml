```python
# DNA sequence
DNA = 'AAAACCCGGT'
```

```python
# reversing DNA
reversed_DNA = DNA[::-1]
```

```python
# Creating dictionary of pairs
pairs = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
```

```python
reverse_complement = ''.join([pairs[i] for i in reversed_DNA])
```

```python
print(reverse_complement)
```
